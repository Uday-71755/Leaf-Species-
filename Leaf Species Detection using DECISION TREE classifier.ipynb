# **Leaf Species Detection using DECISION TREE**

### *Importing required Libraries*

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

### *Load Dataset*

```python
#Iris Dataset
dataset = load_iris()
```

### *Summarize Dataset*

```python
print(dataset.data)
print(dataset.target)
print(dataset.data.shape)
```

### *Segregate Dataset into X(Input/IndependentVariable) & Y(Output/DependentVariable)*

```python
#Segregate the dataset

X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
X
```

```python
Y = dataset.target
Y
```

### *Splitting Dataset into Train & Test*

```python
#Training and Testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.50, random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

### *Finding best max_depth Value*

```python
#Now finding the best max-depth value

accuracy = [] #to store all the max-depth values
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Iterating over max depth values
for i in range(1, 10):
    model = DecisionTreeClassifier(max_depth = i, random_state = 0)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    score = accuracy_score(y_test, pred)
    accuracy.append(score)

#Now plotting between accuracy score and max depth values

plt.figure(figsize=(12, 6))
plt.plot(range(1, 10), accuracy, color="red", linestyle="dashed", marker="o",
         markerfacecolor="blue", markersize=10)
plt.title("Finding best Max_Depth")
plt.xlabel("pred")
plt.ylabel("score")    
```

### *Training*

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy',max_depth = 3, random_state = 0)
model.fit(X_train,y_train)
```

### *Prediction*

```python
#Prediction

y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### *Accuracy Score*

```python
#Computing the accuracy

from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
